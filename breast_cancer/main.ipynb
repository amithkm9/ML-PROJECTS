```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.drop(['Unnamed: 32','id'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.describe()
```

DECISION TREE CLASSIFIER


```python
# Encode the 'diagnosis' column
df['diagnosis_encoded'] = df['diagnosis'].map({'M': 1, 'B': 0})

# Calculate correlations with the encoded column
correlations = df.corr(numeric_only=True)['diagnosis_encoded'].sort_values()

# Display the results
print(correlations)
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x=df['diagnosis'].value_counts().index,y=df['diagnosis'].value_counts().values)
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(20,20))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(['diagnosis'],axis = 1),df['diagnosis'],test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()

dtree.fit(X_train,y_train)
```

```python
y_pred = dtree.predict(X_test)
```

MODEL EVALUATION

```python
print("predicted values",y_pred[:1000])
print("actual values",y_test[:1000])
```

```python
print(dtree.score(X_test,y_test))
```

USING LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
pre = logmodel.predict(X_test)
```

```python
print("predicted values",pre[:10])
print("actual values",y_test[:10])
```

```python
print(logmodel.score(X_test,y_test))
```

FROM THE BOTH MODELS 
IT SHOWS 100% ACCURACY IN BOTH THE MODELS.
WE CAN SAY THAT THE BOTH MODELS ARE EQUAL IN ACCURACCY
